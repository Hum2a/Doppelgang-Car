```python
import os
import requests
from deepface import DeepFace
from werkzeug.utils import secure_filename

# Define the API key and endpoint
API_KEY = "ZxumjjS1ID9tk21k34zCZQ==AlUzWWFvPEaruyWL"
BASE_URL = "https://api.api-ninjas.com/v1/cars"

# Upload folder
UPLOAD_FOLDER = "uploads"
os.makedirs(UPLOAD_FOLDER, exist_ok=True)

def fetch_car_data(make, model, year):
    headers = {"X-Api-Key": API_KEY}
    params = {"make": make, "model": model, "year": year}
    try:
        response = requests.get(BASE_URL, headers=headers, params=params)
        response.raise_for_status()
        return response.json()
    except requests.exceptions.RequestException as e:
        print(f"Error fetching car data: {e}")
        return None
```

```python
from IPython.display import Image, display
import cv2

# Upload the image
image_path = "path_to_your_image.jpg"  # Replace with the path to your test image
display(Image(filename=image_path))  # Show the uploaded image

# Analyze facial features
try:
    features = DeepFace.analyze(img_path=image_path, actions=["age", "gender", "emotion"])
    print(f"Extracted Features: {features}")
except Exception as e:
    print(f"Error analyzing image: {e}")
```

```python
def map_features_to_car(age, gender, emotion):
    if age < 25:
        if gender == "Man":
            return "Ford", "Mustang", 2022
        else:
            return "Volkswagen", "Beetle", 2021
    elif age < 40:
        if emotion == "happy":
            return "Tesla", "Model 3", 2023
        else:
            return "BMW", "3 Series", 2020
    else:
        if emotion == "neutral":
            return "Mercedes-Benz", "E-Class", 2019
        else:
            return "Toyota", "Camry", 2018

# Test the mapping
age = features["age"]
gender = features["gender"]
emotion = features["dominant_emotion"]
make, model, year = map_features_to_car(age, gender, emotion)
print(f"Mapped Car: {make}, {model}, {year}")
```
